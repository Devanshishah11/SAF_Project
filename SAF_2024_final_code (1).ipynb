# Hypertuning

```python
# Train the model
results = model.train(
    data="C:/Users/ds678/OneDrive - University of Sussex/SAF_Project/Instance_seg_labels/SAF_2024_dataset.yaml",
    imgsz=1024,
    epochs=100,
    batch=8,  # Increase batch size if memory allows
    device=0,
    task='segment',
    workers=8,  # Increase number of workers for faster data loading
    optimizer='AdamW',  # Use AdamW optimizer
    cos_lr=True,  # Use cosine learning rate scheduler
    close_mosaic=10,  # Disable mosaic augmentation for final 10 epochs
    amp=True,  # Use mixed precision training
    degrees=45.0,  # Image rotation (+/- deg)
    translate=0.1,  # Image translation (+/- fraction)
    scale=0.5,  # Image scale (+/- gain)
    shear=10.0,  # Image shear (+/- deg)
    perspective=0.0001,  # Image perspective (+/- fraction), range 0-0.001
    flipud=0.5,  # Image flip up-down (probability)
    fliplr=0.5,  # Image flip left-right (probability)
)

# Save the trained model
model.save('droplet_model.pt')
```
